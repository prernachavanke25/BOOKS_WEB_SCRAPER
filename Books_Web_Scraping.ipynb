# Import Library

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Send Request to Website

```python
URL = 'https://books.toscrape.com/'
page = requests.get(URL)
```

# Convert raw HTML code into structured searchable format

```python
soup = BeautifulSoup(page.content, "html.parser")
```

# Extract Books

```python
books_table = soup.find_all("article", class_= "product_pod")
```

# Initialize Empty List

```python
book_titles = []
prices = []
ratings = []
```

# Extract Book Details 

```python
for book in books_table:
    title = book.find("h3").text.strip()
    price = book.find("p", class_="price_color").text.strip()
    rating_tag = book.find("p" , class_="star-rating")
    if rating_tag:
        ratings.append(rating_tag["class"][1])
    else:
        ratings.append("Not Rated")
    book_titles.append(title)
    prices.append(price) 
```

```python
#print(len(book_titles))
#print(len(prices))
#print(len(ratings))
```

# Create DataFrame

```python
df = pd.DataFrame({
    "Book Title": book_titles,
    "Price": prices,
    "Rating": ratings
})
```

# Export to Excel/CSV

```python
df.to_excel('Books.xlsx' , index=False)
print('Excel File')
```
